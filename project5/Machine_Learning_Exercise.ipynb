## Step 1: Specify Prediction Target
Select the target variable, which corresponds to the sales price. Save this to a new variable called <b>y</b>. <br>

```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
housing = pd.read_csv('housing_train.csv')
housing
```

```python
y = housing['SalePrice']
```

## Step 2: Create X
Now you will create a DataFrame called X holding the predictive features.

Since you want only some columns from the original data, you'll first create a list with the names of the columns you want in X.

You'll use just the following columns in the list (you can copy and paste the whole list to save some typing, though you'll still need to add quotes):
- LotArea
- YearBuilt
- 1stFlrSF
- 2ndFlrSF
- FullBath
- BedroomAbvGr
- TotRmsAbvGrd
After you've created that list of features, use it to create the DataFrame that you'll use to fit the model.

```python
feature = ['LotArea','YearBuilt','1stFlrSF','2ndFlrSF','FullBath','BedroomAbvGr','TotRmsAbvGrd']

X = housing[feature]

X
```

## Step 3: Specify and Fit Model¶

- Create a DecisionTreeRegressor and save it iowa_model. Ensure you've done the relevant import from sklearn to run this command.

Then fit the model you just created using the data in X and y that you saved above.

<hr>
<strong>What is a Decision Tree Regressor </strong>
<p>A decision tree regressor is a machine learning model that uses a tree-like structure to predict a continuous target variable . It works by recursively partitioning the data into smaller and smaller subsets based on the features, ultimately fitting a simple model (like the mean or median of the target variable) to each subset</p>
<img src= 'https://miro.medium.com/1*62Am0QdlxCq5Vmt5siR-7Q.png' />

```python
from sklearn.tree import DecisionTreeRegressor

# Define model
iowa_model = DecisionTreeRegressor()

# Fit model
iowa_model.fit(X, y)
```

## Step 4: Make Predictions
Make predictions with the model's predict command using X as the data. Save the results to a variable called predictions.

```python
# Make predictions
predictions = iowa_model.predict(X)

# Show first 5 predictions
print(predictions[:-5])
```

## Step 5: TESTING THE ACCURACY OF PREDICTIONS
Since the sales prices are continuing data or ordinal data. 



```python
from sklearn.metrics import r2_score

r2 = r2_score(y, predictions)
print("R² Score:", r2)
```

<h1>Linear </h1>

```python
from sklearn.linear_model import LinearRegression

# Define model
iowa_model = LinearRegression()

# Fit model
iowa_model.fit(X, y)
```

```python
# Make predictions
predictions = iowa_model.predict(X)

# Show first 5 predictions
print(predictions[:-5])

from sklearn.metrics import r2_score

r2 = r2_score(y, predictions)
print(f"R² Score: {r2: .1%}")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.hist(y, bins=30, alpha=0.6, label='Actual Values')
plt.hist(predictions, bins=30, alpha=0.6, label='Predicted Values')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Distribution of Actual vs. Predicted Values')
plt.legend()
plt.show()

```

<h1>Insight</h1>
<p>The overlapping histograms show how closely the model’s predicted values match the distribution of the actual target values. If the two curves overlap substantially, it indicates that the model captures the overall shape, range, and central tendency of the real data. This means the model is not just predicting random values—it is learning the underlying pattern of the dataset.</p>
